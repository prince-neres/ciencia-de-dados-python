```python
import json
import numpy
import requests

api_key = 'ab249b039cc47d2939d6a12b2dc7664e'

def get_top_artist(api_key):
  url = f"http://ws.audioscrobbler.com/2.0/?method=geo.gettopartists&country=spain&api_key={api_key}&format=json"
  data = requests.get(url).text
  data = json.loads(data)
  tops = []

  #Estou levando em conta o mais popular o mais ouvido
  for i in range(numpy.size(data['topartists']['artist'])):
    tops.append(data['topartists']['artist'][i]['listeners'])
    
    if max(tops) == data['topartists']['artist'][i]['listeners']:
      top = data['topartists']['artist'][i]['name']

  return top

if __name__ == '__main__':
  print(get_top_artist(api_key))
```
