```python
!pip install -U pandasql

import pandas
import pandasql

def select_first_50(filename):

  aadhaar_data = pandas.read_csv(filename)
  
  columns = []
  for i in aadhaar_data.columns:
    new_columns = i.lower().replace(' ', '_')
    columns.append(new_columns)
  
  aadhaar_data.columns = columns

  q = """
    select registrar, enrolment_agency 
    from aadhaar_data
    limit 50
  """
  aadhaar_solution = pandasql.sqldf(q.lower(), locals())
  return aadhaar_solution    

select_first_50('aadhaar-data.csv')
```
