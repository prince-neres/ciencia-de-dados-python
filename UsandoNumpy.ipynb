# Média

```python
array1 = [1,2,3,5,7]                 
array2 = [1,2,3,5,7,9]

def media(num):
  media = sum(num) / len(num)
  return media

media(array1) 
```

# Mediana

```python
array1 = [1,2,3,5,7]                 
array2 = [1,2,3,5,7,9]

def mediana(num):
    meio = len(num) // 2
    num.sort()
    if not len(num) % 2:
        return (num[meio - 1] + num[meio]) / 2.0
    return num[meio]

mediana(array1)
```

# Desvio Padrão

```python
array1 = [1,2,3,5,7]                 
array2 = [1,2,3,5,7,9]

def desvPadrao(num):
  media = sum(num) / len(num)
  soma = 0
  for i in num:
    soma += (i-media)**2 
  return (soma/len(num))**0.5

desvPadrao(array1)
```
