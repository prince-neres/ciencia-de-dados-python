```python
import pandas as pd
from pandas import Series, DataFrame

notas = pd.read_csv('ratings.csv')
filmes = pd.read_csv('movies.csv')
```

```python
# Melhores 
spiderMan = notas.query('movieId==5349')
ironMan = notas.query('movieId==59315')
matrix = notas.query('movieId==2571')

# Piores
anaconda = notas.query('movieId==1499')
mortalKombat = notas.query('movieId==44')
deadPool = notas.query('movieId==187593')
```

```python
# Criando data frame dos 3 filmes que gostei e 3 que não
filmes_nomes = ['Anaconda (1997)', 'Mortal Kombat (1995)', 'Deadpool 2 (2018)',
                'Spider-Man (2002)', 'Iron Man (2008)', 'Matrix, The (1999)']

notas_media = [anaconda.rating.mean(), mortalKombat.rating.mean(), deadPool.rating.mean(),
               spiderMan.rating.mean(), ironMan.rating.mean(), matrix.rating.mean()]
              
like_or_dont = ["Don't like", "Don't like", "Don't like","Like", "Like", "Like"]

filmes_notas = {
    'nome': Series(filmes_nomes),
    'notas': Series(notas_media),
    'like': Series(like_or_dont)
}

df = DataFrame(filmes_notas)
```

```python
# Criando Gráfico
# !pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt

grafico = sns.barplot(x='nome', y='notas', palette="Set1", hue='like', data=df, dodge=False)
grafico.set_title('Notas de Filmes', fontsize='20')
grafico.legend(loc='lower right', fontsize='12')
grafico.tick_params(axis='x', rotation=90)
grafico.set_xlabel('Título do Filme', fontsize='12')
grafico.set_ylabel('Média das Notas', fontsize='12')
grafico.tick_params(labelsize=10)
```

```python
# Salvando gráfico em png
sfig = grafico.get_figure()
sfig.set_size_inches(8, 15)
sfig.savefig('grafico.png', dpi=100)
```
